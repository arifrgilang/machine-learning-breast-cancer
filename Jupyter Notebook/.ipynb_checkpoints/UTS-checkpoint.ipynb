```python
import pandas as pd
import numpy as np

cols = ['Actual',
        'handicapped-infants',
        'water-project-cost-sharing',
        'adoption-of-the-budget-resolution',
        'physician-fee-freeze',
       'el-salvador-aid',
       'religious-groups-in-schools',
       'anti-satellite-test-ban',
       'aid-to-nicaraguan-contras',
       'mx-missile',
       'immigration',
       'synfuels-corporation-cutback',
       'education-spending',
       'superfund-right-to-sue',
       'crime',
       'duty-free-exports',
       'export-administration-act-south-africa']

df = pd.read_csv("house-votes-84.csv", header=None) # baca file csv
df.columns = cols # assign nama kolom
df.head() # baca 5 baris pertama
```

```python
df.replace('?',np.nan, inplace=True) 
fill_mode = lambda col: col.fillna(col.mode()[0]) # isi nilai null dengan modus
df = df.apply(fill_mode, axis=0) 
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # class konversi nilai
new_df = df.apply(le.fit_transform) # ngubah nilai string jadi numeric
new_df.head()
```

```python
df_features = new_df.loc[:,df.columns!='Actual'] # ambil data feature
df_target = new_df[['Actual']] # ambil data target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df_features,df_target,test_size=0.3)

X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
```

## Perceptron

```python
from random import random

def perceptron_train(features, target, maxEpoch, lr):
    #deklarasi variabel awal
    b = 1 #bias
    weight = [random()] * features.shape[1] #weight untuk setiap feature
    y_predict = [0] * features.shape[0] #hasil prediksi untuk setiap baris
    multiply = 0 #penampung hasil perkalian nilai feature dengan weight
    epoch = 0 #nilai epoch
    error = 0
    
    #deklarasi variabel bias dan weight terbaik
    best_b = 1
    best_weight = [0] * features.shape[1]
    best_predict = [0] * features.shape[0]
    lowest_error = np.inf
    
    #Looping untuk setiap barisnya
    while(epoch < maxEpoch):
        error = 0 #nilai error pada satu epoch
        for index,row in features.iterrows():
            multiply = 0
            for i,value in enumerate(row):
                multiply += weight[i] * value #jumlahkan hasil perkalian untuk setiap feature

            Y_in = b + multiply #jumlahkan hasil perkalian dengan bias
            #fungsi aktivasi binary step
            if(Y_in >= 0): #jika nilai penjumlahan positif, anggap prediksi adalah positif
                y_predict[index] = 1
            else: #jika nilai penjumlahan negatif, anggap prediksi adalah negatif
                y_predict[index] = 0

            if(y_predict[index] != target.iloc[index,0]): #jika terdapat perbedaan hasil prediksi dengan target asli, perbaiki weight dan bias
                for i,value in enumerate(row): #perbaikan weight
                    weight[i] += lr * (target.iloc[index,0] - y_predict[index]) * value
                b += lr * (target.iloc[index,0] - y_predict[index]) #perbaikan bias
                error += 1 #tambahkan error
                
            if(index==features.shape[0]-1): #jika seluruh baris telah dihitung, tambahkan nilai epoch
                epoch += 1
                
        if(error < lowest_error): #penyimpanan nilai bias, weight, hasil prediksi, dan error terbaik
            best_b = b
            best_weight = weight.copy()
            best_predict = y_predict.copy()
            best_epoch = epoch
            lowest_error = error
            
        print("Accuracy Epoch Ke ",epoch," : ",str(round(1-error/features.shape[0],3)*100)[0:4],"%") #print akurasi setiap epochnya
    return best_weight, best_b, best_predict, best_epoch
```

```python
model_weight, model_bias, model_predict, model_epoch = perceptron_train(df_features,df_target, 50, 0.1)
```

```python
from random import random

def perceptron_predict(features, weight, b):
    #deklarasi variabel awal
    y_predict = [0] * features.shape[0] #hasil prediksi untuk setiap baris
    multiply = 0 #penampung hasil perkalian nilai feature dengan weight
    
    for index,row in features.iterrows():
        multiply = 0
        for i,value in enumerate(row):
            multiply += weight[i] * value #jumlahkan hasil perkalian untuk setiap feature

        Y_in = b + multiply #jumlahkan hasil perkalian dengan bias
        #fungsi aktivasi binary step
        if(Y_in >= 0): #jika nilai penjumlahan positif, anggap prediksi adalah positif
            y_predict[index] = 1
        else: #jika nilai penjumlahan negatif, anggap prediksi adalah negatif
            y_predict[index] = 0
    return y_predict
```
